```python
import numpy as np
```

```python
class TicTacToe():
    
    def __init__(self,epochs):
        
        self.state_index = {}
        self.state_reward = {}
        self.index_state = {}
        self.num_states = None
        self.num_actions = 9
        self.player = 1
        self.state = 0
        
        self.state_reward[tuple(np.ones(9) * -1)] = 0
        
        for i in range(epochs):
    
            board = np.ones(9) * -1
            player = 1
            while(True):
                board = self.player_move(player,board)
                win = self.get_reward(player,board)
                k = tuple(board)
                if(win):             
                    self.state_reward[k] = player*2 - 1
                    break   
                self.state_reward[k] = 0
                if(len(np.where(board == -1)[0]) == 0):
                    break
                player = int(not(player)) 
        
        self.create_state_id()
        
    def available_actions(self):
        
        board = np.array(self.index_state[self.state])
        return np.where(board == -1)[0]
                
    def player_move(self,player,board):
    
        pos = np.where(board == -1)[0]
        index = np.random.choice(pos)
        board[index] = player
        return board
    
    def get_reward(self,player,_board):

        board = _board.reshape(3,3)
        row,col = np.where(board == player)
        win = 0
        for i in range(3):

            if(np.sum(row == i) == 3):
                win = player * 2 -1
                break

            if(np.sum(col == i) == 3):
                win = player * 2 -1
                break

        if(np.sum(board[[0,1,2],[0,1,2]] == player) == 3):
            win = player * 2 -1

        if(np.sum(board[[2,1,0],[0,1,2]] == player) == 3):
            win = player * 2 -1

        return win
    
            
    def create_state_id(self):
        
        key = 0
        for i in list(self.state_reward.keys()):

            self.state_index[i] = key
            key+=1

        self.num_states = len(self.state_index)
        
        for i in list(self.state_index.keys()):

            self.index_state[self.state_index[i]] = i
    
    def make_move(self,action):
        
        board = np.array(self.index_state[self.state])
        board[action] = self.player
        reward = self.state_reward[tuple(board)]
        self.state = self.state_index[tuple(board)]
        self.player = int(not self.player)
        return self.state,reward
        
   
    def get_state_reward(self):
        
        return self.state_reward
    
    def get_state_index(self):
        
        return self.state_index
    
    def get_index_state(self):
        
        return self.index_state
    
    def print_state(self):
        
        instance = self.index_state[self.state]
        print(np.array(instance).reshape(3,3) + 1)
        
    def reset_env(self):
        
        self.state = 0
        self.player = 1
        return self.state    
    
    def play_game(self,Q,player):
        
        self.reset_env()
        s = self.state
        while(True):
        
            if (player == 1):
                self.print_state()
                a_player = input("choose move")
                s,r = self.make_move(int(a_player))
                
                if(r != 0 ):
                    self.print_state()
                    break    
                actions = self.available_actions()
                
                if(len(actions) == 0):
                    break
                    
                a_bot = actions[np.argmin(Q[s,list(actions)])]
                s,r = self.make_move(a_bot)
                if(r != 0 ):
                    self.print_state()
                    break 
                
            else :
                    
                    self.print_state()
                    actions = self.available_actions()
                    a_bot = actions[np.argmax(Q[s,list(actions)])]
                    s,r = self.make_move(a_bot)
                    if(r != 0 ):
                        self.print_state()
                        break
                        
                    self.print_state()
                    
                    a_player = input("choose move")
                    s,r = self.make_move(int(a_player))
                    if(r != 0 ):
                        self.print_state()
                        break
                    
                
        
```

```python
env = TicTacToe(20000)
```

```python
Q = np.zeros([env.num_states,env.num_actions])
```

```python
# Set learning parameters
lr = .7
exp = 0.7
y = 0.8
num_episodes = 10000000

#create lists to contain total rewards and steps per episode
#jList = []
rList = []
for i in range(num_episodes):
    #Reset environment and get first new observation
    s = env.reset_env()
    rAll = 0
    j = 0
    #The Q-Table learning algorithm
    while j < 99:
        j+=1
        #Choose an action by greedily (with noise) picking from Q table
        actions = env.available_actions()
        if (len(actions) == 0):
            break
        p = np.random.rand(1)
        if( p< exp):
            
            a = np.random.choice(actions)
        else:
            
            if(env.player):
                
                a = actions[np.argmax(Q[s,list(actions)])]
            
            else:
                
                a = actions[np.argmin(Q[s,list(actions)])]
                
        #Get new state and reward from environment
        s1,r = env.make_move(a)
        #Update Q-Table with new knowledge
        if (env.player):
            Q[s,a] = Q[s,a] + lr*(r + y*np.max(Q[s1,:]) - Q[s,a])
        else:
            Q[s,a] = Q[s,a] + lr*(r + y*np.min(Q[s1,:]) - Q[s,a])
        rAll += r
        s = s1
        if (r == -1 or r == 1):
            break
        
    #jList.append(j)
    rList.append(rAll)

        
```
